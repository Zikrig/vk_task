```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GroupShuffleSplit
from sklearn.metrics import classification_report
df = pd.read_csv('intern_task.csv')

splitter = GroupShuffleSplit(test_size=.2, n_splits=4, random_state = 7)
splitt = splitter.split(df, groups=df['query_id'])
train_inds, test_inds = next(splitt)

# print(train_inds)
train = df.iloc[train_inds]
test = df.iloc[test_inds]

y_train  = df.iloc[train_inds]['rank']
y_test  = df.iloc[test_inds]['rank']

X = df
X = X.drop(columns=['rank', 'query_id'])
X_train = X.iloc[train_inds]
X_test = X.iloc[test_inds]
```

```python

from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import ndcg_score

etc = ExtraTreesClassifier(n_estimators=4)
etc.fit(X_train, y_train)
score = etc.score(X_test, y_test)
print(score)

res = etc.predict(X_test)
print(res)
s = classification_report(y_test, res)
print(s)

ndc = ndcg_score(y_test, res)
print(ndc)
```
